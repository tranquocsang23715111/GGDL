# Câu 1: So sánh MST và Shortest Path

## 1. Mục tiêu
- MST: Kết nối tất cả các đỉnh với tổng trọng số nhỏ nhất
- Shortest Path: Tìm đường đi ngắn nhất từ một đỉnh đến đỉnh khác

## 2. Input / Output
- MST:
  - Input: Graph vô hướng, có trọng số
  - Output: Cây gồm n-1 cạnh
- Shortest Path:
  - Input: Graph (có thể có hướng), có trọng số
  - Output: Khoảng cách / đường đi

## 3. Thuật toán
- MST: Prim, Kruskal
- Shortest Path: Dijkstra, Bellman-Ford

## 4. Khác biệt chính
- MST: Bao phủ toàn bộ graph
- Shortest Path: Chỉ quan tâm đường giữa các node

# Câu 2: Thuật toán Prim

## Ý tưởng
- Bắt đầu từ một đỉnh
- Mỗi bước chọn cạnh nhỏ nhất nối cây với đỉnh chưa chọn

## Các bước
1. Chọn node bắt đầu
2. Đưa các cạnh kề vào priority queue
3. Lặp:
   - Lấy cạnh nhỏ nhất
   - Nếu node chưa chọn → thêm vào cây

## Độ phức tạp
- O(E log V) (dùng heap)

# Câu 3: Thuật toán Kruskal

## Cách hoạt động
- Sắp xếp cạnh theo trọng số tăng dần
- Lần lượt chọn cạnh nhỏ nhất không tạo chu trình

## Union-Find
- Dùng để kiểm tra chu trình
- Nếu 2 node cùng tập → bỏ qua
- Nếu khác → union

## Khi dùng Kruskal
- Graph thưa (ít cạnh)
- Dễ cài đặt

# Câu 4: Thuật toán Dijkstra

## Điều kiện áp dụng
- Graph có trọng số không âm

## Vì sao không dùng cạnh âm?
- Thuật toán greedy
- Khi chọn rồi không quay lại → sai nếu có cạnh âm

## Tối ưu
- Dùng priority queue (heap)
- Giảm từ O(V^2) → O(E log V)

# Câu 5: So sánh thuật toán

## Dijkstra vs BFS
- BFS: graph không trọng số
- Dijkstra: graph có trọng số

## Dijkstra vs Bellman-Ford
- Dijkstra: nhanh, không xử lý cạnh âm
- Bellman-Ford: chậm hơn, xử lý được cạnh âm

```python
#prim
import heapq

def prim(graph, start):
    visited = set()
    pq = [(0, start)]
    total = 0
    edges = []

    while pq:
        w, u = heapq.heappop(pq)
        if u in visited:
            continue

        visited.add(u)
        total += w

        for v, weight in graph[u]:
            if v not in visited:
                heapq.heappush(pq, (weight, v))
                edges.append((u, v, weight))

    return total, edges

# test
graph = {
    0: [(1, 2), (3, 6)],
    1: [(0, 2), (2, 3), (3, 8)],
    2: [(1, 3)],
    3: [(0, 6), (1, 8)]
}

print(prim(graph, 0))
```

```python
#kruskal
def find(p, x):
    if p[x] != x:
        p[x] = find(p, p[x])
    return p[x]

def union(p, x, y):
    p[find(p, x)] = find(p, y)

def kruskal(n, edges):
    edges.sort(key=lambda x: x[2])
    p = list(range(n))
    mst = []
    total = 0

    for u, v, w in edges:
        if find(p, u) != find(p, v):
            union(p, u, v)
            mst.append((u, v, w))
            total += w

    return total, mst

# test
edges = [(0,1,2),(1,2,3),(0,2,4)]
print(kruskal(3, edges))
```

```python
#dijkstra
import heapq

def dijkstra(graph, start):
    dist = {v: float('inf') for v in graph}
    dist[start] = 0

    pq = [(0, start)]

    while pq:
        d, u = heapq.heappop(pq)

        if d > dist[u]:
            continue

        for v, w in graph[u]:
            if dist[v] > dist[u] + w:
                dist[v] = dist[u] + w
                heapq.heappush(pq, (dist[v], v))

    return dist

# test
graph = {
    0: [(1,4),(2,1)],
    1: [(3,1)],
    2: [(1,2),(3,5)],
    3: []
}

print(dijkstra(graph, 0))
```

# Câu 10: Mạng điện (Minimum Spanning Tree)

Một thành phố có 6 trạm điện được đánh số từ 0 → 5. 
Chi phí nối dây giữa các trạm được cho bởi danh sách cạnh dưới đây.

Yêu cầu:
1. Xây dựng cây khung nhỏ nhất (MST)
2. In ra:
   - Tổng chi phí
   - Danh sách cạnh trong MST

Dữ liệu:
edges = [
    (0,1,4),
    (0,2,4),
    (1,2,2),
    (2,3,3),
    (2,5,2),
    (2,4,4),
    (3,4,3),
    (5,4,3)
]

```python
edges = [
    (0,1,4),(0,2,4),(1,2,2),
    (2,3,3),(2,5,2),(2,4,4),
    (3,4,3),(5,4,3)
]

total, mst = kruskal(6, edges)

print("Total cost:", total)
print("MST:", mst)
```

# Câu 11: Google Maps (Shortest Path)

Cho graph biểu diễn hệ thống đường đi giữa các địa điểm.

Yêu cầu:
1. Tìm khoảng cách ngắn nhất từ node 0 đến tất cả các node
2. In kết quả dưới dạng dictionary

Dữ liệu:
graph = {
    0: [(1,4),(2,1)],
    1: [(3,1)],
    2: [(1,2),(3,5)],
    3: []
}

```python
import heapq

def dijkstra(graph, start):
    dist = {v: float('inf') for v in graph}
    dist[start] = 0
    pq = [(0, start)]

    while pq:
        d, u = heapq.heappop(pq)

        if d > dist[u]:
            continue

        for v, w in graph[u]:
            if dist[v] > dist[u] + w:
                dist[v] = dist[u] + w
                heapq.heappush(pq, (dist[v], v))

    return dist

graph = {
    0: [(1,4),(2,1)],
    1: [(3,1)],
    2: [(1,2),(3,5)],
    3: []
}

print(dijkstra(graph, 0))
```

# Câu 13: Graph có cạnh âm

Cho graph có cạnh âm như sau:

Yêu cầu:
1. Giải thích vì sao không dùng Dijkstra
2. Dùng Bellman-Ford để tìm khoảng cách ngắn nhất từ node 0

Dữ liệu:
edges = [
    (0,1,1),
    (1,2,-1),
    (2,3,-1)
]

```python
def bellman_ford(edges, n, start):
    dist = [float('inf')] * n
    dist[start] = 0

    for _ in range(n-1):
        for u, v, w in edges:
            if dist[u] + w < dist[v]:
                dist[v] = dist[u] + w

    return dist

edges = [(0,1,1),(1,2,-1),(2,3,-1)]

print(bellman_ford(edges, 4, 0))
```
