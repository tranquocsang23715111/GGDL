### Bài 3. (3.0 đ)
- Sinh graph đơn vô hướng có $10$ đỉnh (đánh số từ $1 \to 10$), có $30$ cạnh không trọng số.
- Tính số thành phần liên thông trong graph.
- Kiểm tra xem graph có lưỡng phân hay không.
- Mô tả cây DFS tạo thành khi duyệt graph từ đỉnh $1.$
- Liệt kê các khớp, cầu của graph.
- Vẽ graph và tô đỏ cho khớp, tô xanh cho cầu.

```python
import networkx as nx
import matplotlib.pyplot as plt
import random

# 1. Sinh graph 10 đỉnh, 30 cạnh
nodes = list(range(1, 11))
G = nx.Graph()
G.add_nodes_from(nodes)

# sinh tất cả cạnh có thể có
all_edges = [(i, j) for i in nodes for j in nodes if i < j]

# chọn ngẫu nhiên 30 cạnh
edges = random.sample(all_edges, 30)
G.add_edges_from(edges)

print("Số đỉnh:", G.number_of_nodes())
print("Số cạnh:", G.number_of_edges())

# 2. Số thành phần liên thông
num_cc = nx.number_connected_components(G)
print("Số thành phần liên thông:", num_cc)

# 3. Kiểm tra lưỡng phân
is_bipartite = nx.is_bipartite(G)
print("Graph có lưỡng phân không:", is_bipartite)

# 4. DFS tree từ đỉnh 1
dfs_tree = nx.dfs_tree(G, source=1)
print("Cây DFS (các cạnh):")
print(list(dfs_tree.edges()))

# 5. Khớp (articulation points) và cầu (bridges)
articulation_points = list(nx.articulation_points(G))
bridges = list(nx.bridges(G))

print("Đỉnh khớp:", articulation_points)
print("Cạnh cầu:", bridges)

# 6. Vẽ graph
pos = nx.spring_layout(G, seed=42)

plt.figure(figsize=(8,6))

# vẽ node
nx.draw(G, pos, with_labels=True, node_color='lightgray')

# tô đỏ đỉnh khớp
nx.draw_networkx_nodes(G, pos, nodelist=articulation_points, node_color='red')

# tô xanh cạnh cầu
nx.draw_networkx_edges(G, pos, edgelist=bridges, edge_color='blue', width=2)

plt.title("Graph với articulation points (đỏ) và bridges (xanh)")
plt.show()
```

### Bài 2. (2.0 đ)
- Sinh một graph đơn vô hướng có $20$ đỉnh, $100$ cạnh và các cạnh có trọng số nguyên từ $1 \to 20.$
- Xây dựng ma trận khoảng cách ngắn nhất giữa tất cả các cặp đỉnh bằng thuật toán Floy-Warshall.
- Mô tả đường đi ngắn nhất giữa hai đỉnh $1 \to 20$ bằng thuật toán Dijkstra.
- Tìm cây khung có trọng số nhỏ nhất.

```python
import networkx as nx
import matplotlib.pyplot as plt
import random
import numpy as np

# 1. Sinh graph 20 đỉnh, 100 cạnh, trọng số 1-20
nodes = list(range(1,21))
G = nx.Graph()
G.add_nodes_from(nodes)

# tất cả các cạnh có thể có
all_edges = [(i,j) for i in nodes for j in nodes if i<j]

# chọn 100 cạnh ngẫu nhiên
edges = random.sample(all_edges, 100)

# gán trọng số ngẫu nhiên từ 1-20
weighted_edges = [(u,v,random.randint(1,20)) for u,v in edges]
G.add_weighted_edges_from(weighted_edges)

print("Số đỉnh:", G.number_of_nodes())
print("Số cạnh:", G.number_of_edges())

# 2. Ma trận khoảng cách ngắn nhất bằng Floyd-Warshall
dist_matrix = dict(nx.floyd_warshall(G, weight='weight'))
# chuyển sang numpy array để dễ nhìn
dist_array = np.zeros((20,20))
for i in range(1,21):
    for j in range(1,21):
        dist_array[i-1,j-1] = dist_matrix[i][j]
print("Ma trận khoảng cách ngắn nhất (Floyd-Warshall):\n", dist_array)

# 3. Đường đi ngắn nhất từ 1->20 bằng Dijkstra
shortest_path = nx.dijkstra_path(G, source=1, target=20, weight='weight')
shortest_length = nx.dijkstra_path_length(G, source=1, target=20, weight='weight')
print("Đường đi ngắn nhất từ 1 -> 20:", shortest_path)
print("Độ dài đường đi:", shortest_length)

# 4. Tìm cây khung nhỏ nhất (MST)
mst = nx.minimum_spanning_tree(G, weight='weight')
print("Cạnh của cây khung nhỏ nhất (MST):")
for u,v,d in mst.edges(data=True):
    print(f"({u},{v}) trọng số={d['weight']}")

# 5. Vẽ graph
plt.figure(figsize=(10,6))
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True, node_color='lightgray')
nx.draw_networkx_edges(G, pos, edgelist=mst.edges(), edge_color='red', width=2)
plt.title("Graph với MST (đỏ)")
plt.show()
```

- Thực hiện thuật toán pagerank, lấy hằng số $\alpha = 0.85$, in ra ma trận khởi tạo và bảng sắp thứ tự tên của các tên khủng bố cùng rank của họ.

*Tham khảo đoạn code bên dưới.*

```python
edge_list = np.loadtxt('terrorist.pairs', dtype=str)
node_name_dict = dict(np.loadtxt('terrorist.names', dtype=str))
```

```python
edge_list = [(node_name_dict[source], node_name_dict[target]) for source, target in edge_list]
edge_list[:10]
```

```python
terrorist_net = nx.DiGraph()
terrorist_net.name = 'terrorist network 911'
terrorist_net.add_edges_from(edge_list)
print("Số đỉnh:", terrorist_net.number_of_nodes())
print("Số cạnh:", terrorist_net.number_of_edges())
print("10 cặp tên đầu tiên:", list(terrorist_net.edges())[:10])
```

```python
nx.draw(terrorist_net, with_labels = True)
```

```python
alpha = 0.85
pr = nx.pagerank(terrorist_net, alpha)
print(nx.google_matrix(terrorist_net, alpha))
```
